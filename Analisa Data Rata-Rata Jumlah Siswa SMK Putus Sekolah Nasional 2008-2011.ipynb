# Deskripsi Data

Data ini mengenai jumlah siswa/i SMK di 33 provinsi yang putus sekolah dari tahun 2008 hingga 2011. Data dalam format .csv dengan nama file "pocessed-data-putus-sekolah-smk-2008-2011.csv". Memiliki 5 kolom yang masing-masing berisi data 'kode_provinsi', 'nama_provinsi', 'tahun', 'jumlah_putus_sekolah', dan
'persen_putus_sekolah'. Analisa kali ini ingin mencari rata-rata siswa/i SMK yang putus sekolah secara nasional setiap tahun dari 2008-2011.

```python
import pandas as pd
```

```python
data_putus_sekolah = pd.read_csv('processed-data-putus-sekolah-smk-2008-2011.csv', delimiter = ',', encoding = "ISO-8859-1") #UTF-88
```

```python
data_putus_sekolah
```

```python
data_putus_sekolah.columns
```

# Statistik Tentang Data
Statistik akan mencari data rata-rata nasional setiap tahunnya dari 208-2011.

```python

data_putus_sekolah.groupby('tahun').mean()
```

# Visualisasi Data

```python
ax= data_putus_sekolah.groupby('tahun').mean()['jumlah_putus_sekolah']
ax= ax.plot(kind='bar', title ="Jumlah Rata-Rata Putus Sekolah Nasional Berdasarkan Tahun",figsize=(10,5),legend=False, fontsize=18)
ax.set_xlabel("tahun", fontsize=10)
ax.set_ylabel("jumlah_putus_sekolah", fontsize=10)
```

# Narasi Analisis
Hasil analisis terlihat di visualisasi, di tahun 2008 menuju 2009, terjadi peningkatan rata-rata jumlah siswa/i SMK yang putus sekolah dalam lingkup nasional, yaitu dari 2849.454545 ke 5176.727273. 2010 terjadi penurunan ke 2989.090909 dan meningkat lagi ke 3781.575758 pada 2011, meskipun masih lebih rendah dari rata-rata nasional di tahun 2009.
